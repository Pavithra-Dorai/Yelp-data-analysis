```python
# Load the JSON data
with open("/kaggle/input/yelp-dataset/yelp_academic_dataset_checkin.json") as f:
    data_checkin = [json.loads(line) for line in f]

# Convert the JSON data to a DataFrame
checkin_df = pd.DataFrame(data_checkin)
```

```python
# Split the 'date' column on commas and expand the results into separate rows
checkin_df = checkin_df.assign(date=checkin_df['date'].str.split(',')).explode('date')

# Strip leading/trailing whitespace
checkin_df['date'] = checkin_df['date'].str.strip()

# Now try converting to datetime
checkin_df['date'] = pd.to_datetime(checkin_df['date'])
```

```python
checkin_df
```

```python
checkin_df.to_sql(name = "checkin", con = engine, if_exists='replace')
```
